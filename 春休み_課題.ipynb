```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
student_mat = pd.read_csv('~/Downloads/student/student-mat.csv', delimiter = ';')
```

```python
print(student_mat)
```

```python
print("Shape of student_mat: {}".format(student_mat.shape))
```

```python
X = student_mat.drop(['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'G3'], axis=1)
# X = student_mat.drop(["G3"], axis=1)
X = X.replace('yes', 1)
X = X.replace('no', 0)
y = student_mat['G3']
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
```

```python
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
for x, t in zip(X, tree.feature_importances_):
    print("{}:{:.2f}% \n".format(x, t*100))
```
